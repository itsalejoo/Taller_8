crear un n-darray de dos dimensiones con la estatura de cada estudiante
-calcular la media, mediana, desviacion estandar
-graficar la data de estatura del grupo. 

```python
import numpy as np
estatura=np.array([[1.88, 1.60, 1.60, 1.60, 1.60, 1.60 , 1.65, 1.59], 
                 [1.75, 1.54, 1.72, 1.74, 1.80, 1.76, 1.76, 1.69]])
estatura
```

```python
media= np.mean(estatura)
media
```

```python
mediana=np.median(estatura)
mediana
```

```python
desviacion= np.std(estatura)
desviacion
```

```python
import matplotlib.pyplot as plt
#simular un conjunto de datos que representen la elevación en un terreno.
#crear un conjunto de datos de prueba 
estatura=np.array([[1.88, 1.60, 1.60, 1.60, 1.60, 1.60 , 1.65, 1.59], 
                  [1.75, 1.54, 1.72, 1.74, 1.80, 1.76, 1.76, 1.69]])

plt.hist(estatura,bins=3)
plt.xlabel("Estatura (m)")
plt.ylabel("frecuencia")
plt.title("estatura del grupo POO")
plt.show()
```
